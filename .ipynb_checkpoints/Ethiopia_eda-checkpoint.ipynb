```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/ethiopia.csv")

df["Country"] = "Ethiopia"

df.head()
```

```python
df["Date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df["Month"] = df["Date"].dt.month
```

```python
df.replace(-999, np.nan, inplace=True)
```

```python
df.replace(-999, np.nan, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
monthly_temp = df.groupby("Month")["T2M"].mean()

plt.figure(figsize=(10,5))
plt.plot(monthly_temp, marker="o")
plt.title("Monthly Average Temperature (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid()
plt.show()
```

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()

plt.figure(figsize=(10,5))
plt.bar(monthly_rain.index, monthly_rain.values)
plt.title("Monthly Total Rainfall (Ethiopia)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["T2M"], df["RH2M"], alpha=0.4)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df["PRECTOTCORR"], bins=30)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")
plt.show()
```

```python
def zscore(df, col):
    return (df[col] - df[col].mean()) / df[col].std()

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z_scores = df[cols].apply(lambda x: abs((x - x.mean()) / x.std()))
outliers = (z_scores > 3).any(axis=1)

print("Outliers:", outliers.sum())
```

```python
df_outliers = df[outliers]
print(df_outliers.head())
print(df_outliers.shape)
```

```python
cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z_scores = abs((df[cols] - df[cols].mean()) / df[cols].std())

for col in cols:
    print(col, (z_scores[col] > 3).sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.boxplot(data=df[cols])
plt.xticks(rotation=45)
plt.show()
```

```python
cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

for col in cols:
    upper = df[col].mean() + 3 * df[col].std()
    lower = df[col].mean() - 3 * df[col].std()
    df[col] = df[col].clip(lower, upper)
```

```python
weather_cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

df[weather_cols] = df[weather_cols].ffill()

threshold = int(df.shape[1] * 0.7)
df = df.dropna(thresh=threshold)
```

```python
df.to_csv("data/ethiopia_clean.csv", index=False)
```
